## Training and Testing Data

```python
y=customer['Yearly Amount Spent']
```

```python
customer.columns
```

```python
x=customer[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
# Splitting the data using sklearn in x_train, x_test, y_train, y_test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

## Training the Model

```python
# Importing the Linear Regression model from sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Creating the instance for linear Regression model named lm
lm=LinearRegression()
```

```python
# Fit/Train the data for implementing thr model
lm.fit(x_train,y_train)
```

```python
# Estimates the coefficient of linear regression model
print('Coefficient: \n', lm.coef_)
```

## Predicting Test Data

```python
# Predicting the x_test values for the scatter plot using lm.predict() function
predictions=lm.predict(x_test)
```

```python
# Create a scatterplot of the real test values versus the predicted values. 
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## Evaluating the Model

```python
# Evaluating the model by calculating Mean Absolute Error, Mean Squared Error, Root Mean Squared Error, Mean Absolute Percentage Error
from sklearn import metrics
print('MAE: ', metrics.mean_absolute_error(y_test,predictions))
print('MSE: ', metrics.mean_squared_error(y_test,predictions))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
print('MAPE: ', metrics.mean_absolute_percentage_error(y_test,predictions))
print('R2 Score: ', metrics.r2_score(y_test,predictions))
```

```python
# Plotting the histogram using seaborn distplot to check the normal distribution 
sns.distplot((y_test-predictions),bins=50);
```

```python
# Using seaborn heatmap to plot correlation matrics as a heat map between different variables
sns.heatmap(customer.corr(),annot=True)
```
